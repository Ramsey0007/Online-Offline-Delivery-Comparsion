```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
# Example code for a histogram
dataset = pd.read_csv("/content/jobs_in_data.csv")
sns.histplot(data=dataset,x='salary_in_usd')
plt.show()
```

```python
# Example code for a line graph
sns.lineplot(data=dataset,x='salary_currency',y='salary')
plt.title('Salary_Description')
plt.xlabel('salary_currency')
plt.ylabel('salary')
plt.show()
```

```python
fig = px.scatter(dataset, x='salary', y='salary_in_usd')
fig.show()
```
